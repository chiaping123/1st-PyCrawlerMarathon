## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
jsn_data=json.loads(response)
print(jsn_data)
```

```python
print("共:",len(jsn_data),"筆資料")
print("每一筆資料包含以下欄位:",jsn_data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for data in jsn_data:
    print("標題:",data['title'])
    print("貼文時間:",data['createdAt'])
    print("留言人數:",data['commentCount'])
    print("按讚人數:",data['likeCount'])
    print()
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
#熱門
new_list_comment=[]
new_list_like=[]
for data in jsn_data:
    #comment_cnt = data['commentCount']
    new_list_comment.append(data['commentCount'])
    new_list_like.append(data['likeCount'])
    #print(type(comment_cnt))
    ##like_cnt = data['likeCount']
    ##new_list.append(int_price)
print("熱門  平均留言人數:",sum(new_list_comment)/len(new_list_comment))
print("熱門  平均按讚人數:",sum(new_list_like)/len(new_list_comment))
```

```python
# 非熱門
import requests
r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2 = r2.text

response2
```

```python
import json
jsn_data2=json.loads(response2)
print("共:",len(jsn_data2),"筆資料")
print("每一筆資料包含以下欄位:",jsn_data2[0].keys())
```

```python
for data2 in jsn_data2:
    print("標題:",data2['title'])
    print("貼文時間:",data2['createdAt'])
    print("留言人數:",data2['commentCount'])
    print("按讚人數:",data2['likeCount'])
    print()
```

```python
new_list_comment2=[]
new_list_like2=[]
for data2 in jsn_data2:
    new_list_comment2.append(data2['commentCount'])
    new_list_like2.append(data2['likeCount'])
print("非熱門 平均留言人數:",sum(new_list_comment2)/len(new_list_comment2))
print("非熱門 平均按讚人數:",sum(new_list_like2)/len(new_list_comment2))
```
