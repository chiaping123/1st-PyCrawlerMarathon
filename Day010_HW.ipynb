```python
print('BeautifulSoup')
print('=======================================')
import requests
from bs4 import BeautifulSoup 
url = 'https://www.dcard.tw/f'
res = requests.get(url)
print(res.text[0:3000])
soup = BeautifulSoup(res.text)
print('=======================================')
print(soup.title)
```

Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
#pip install grab
#pip install PyQuery
res = Grab().go(url)
print("Grab + PyQuery")
print('=====================')
print(res.body[0:100])
doc = pq(res.body.decode('utf-8'))
title = doc('title')
print('================title==================')
print(title)
divs = doc('div')
for div in divs.items():
    print('================div===================')
    print(div)
```
