```python
import requests
import re
from bs4 import BeautifulSoup
```

### 先搜尋全部的電影代號(ID)資訊

```python
# 查看目前上映那些電影，並擷取出其ID資訊
url = 'https://movies.yahoo.com.tw/'
resp = requests.get(url)
resp.encoding = 'utf-8'
soup = BeautifulSoup(resp.text, 'lxml')
#print(soup)
#print('-1------------------------------------------------------')
#print('-1------------------------------------------------------')
html = soup.find("select", attrs={'name':'movie_id'})
#print(html)
#print('-2------------------------------------------------------')
#print('-2------------------------------------------------------')
movie_item = html.find_all("option", attrs={'data-name':re.compile('.*')})
#print(movie_item)
#print('-3------------------------------------------------------')
#print('-3------------------------------------------------------')
for p in movie_item:
    #print(p)
    #print(p["data-name"])
    #print(p["value"])
    print("Movie: %s, ID: %s" % (p["data-name"], p["value"]))
```

### 指定你想要觀看的放映地區，查詢有上映電影的場次日期

```python
# 指定放映地區
area_id = 20
```

```python
# 向網站發送請求
url = 'https://movies.yahoo.com.tw/movietime_result.html'
payload = {'movie_id':str(movie_id), 'area_id':str(area_id)}
#print('-1------------------------------------------------------')
#print(payload)
#https://movies.yahoo.com.tw/movietime_result.html?movie_id=10527&area_id=20
resp = requests.get(url, params=payload)
resp.encoding = 'utf-8'
#print('-2------------------------------------------------------')
#print(resp)
soup = BeautifulSoup(resp.text, 'lxml')
#print('-3------------------------------------------------------')
#print(soup)
movie_date = soup.find_all("label", attrs={'for':re.compile("date_[\d]")})
#print('-4------------------------------------------------------')
#print(movie_date)
#re.compile("date_[\d]") 不確定如何使用
# 列印播放日期
for date in movie_date:
    #print('-5------------------------------------------------------')
    #print(date)
    #print(date.p.string)
    #print(date.h3.string)
    print("時間有: %s %s" % (date.p.string, date.h3.string))
```

### 最後指定觀看的日期，查詢並列印出放映的電影院、放映類型(數位、3D、IMAX 3D...)、放映時間等資訊。

```python
# 選定要觀看的日期
date = "2020-03-01"
```

```python
# 向網站發送請求，獲取上映的電影院及時間資訊
url = "https://movies.yahoo.com.tw/ajax/pc/get_schedule_by_movie"
payload = {'movie_id':str(movie_id),
           'date':date,
           'area_id':'',   #str(area_id)
           'theater_id':'',
           'datetime':'',
           'movie_type_id':''}
#-1------------------------------------------------------')
#print(area_id)
resp = requests.get(url, params=payload)
#print('-2------------------------------------------------------')
#print(resp)  # 若有需要，列印出json原始碼
soup = BeautifulSoup(resp.json()['view'], 'lxml')
#print('-3------------------------------------------------------')
#print(soup)  
html = soup.find_all("ul", attrs={'data-theater_name':re.compile(".*")})
print('-4------------------------------------------------------')
print(html) 
```

```python
for lists in html:
    #print('-1------------------------------------------------------')
    #print(type(lists))
    #print(lists.a.text)
    #print("電影院名稱:{0}".format(lists.a.text,))
    mTypes = lists.find_all('input',{'class':'gabtn'})
    #print('-2------------------------------------------------------')
    #print(mTypes)    
    for mType in mTypes:
        #print('-3------------------------------------------------------')
        #print(mType['data-movie_title']) 
        #print(mType['data-movie_type'])
        #print(mType['value'])
        print("影片名稱:{0} 影片放映類型:{1} 時間表:{2}".format(mType['data-movie_title'],mType['data-movie_type'],mType['value']))
```
